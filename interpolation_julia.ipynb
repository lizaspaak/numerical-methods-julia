# Interpolation with Julia

## Objective
Study polynomial interpolation using Julia and visualize interpolation results.

```julia
using Plots
```

```julia
x = [1, 2, 3, 4, 5]

y = [2, 4, 5, 4, 5]
```

```julia
plot(x, y,
    marker = :circle,
    linewidth = 2,
    title = "Interpolation Data",
    xlabel = "x",
    ylabel = "y")
```

## Conclusion

This project introduces interpolation visualization and scientific plotting using Julia.
